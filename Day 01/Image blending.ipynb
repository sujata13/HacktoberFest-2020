```python
import numpy as np
import cv2
%matplotlib inline
#Comment: The purpose of the above line is to display matplotlib plots inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img_water = mpimg.imread('water.jpg')
resized = cv2.resize(img_water, (500, 500))
plt.imshow(resized)
plt.title('Image')
plt.show()
```

```python
img_girl = mpimg.imread('girl.jpg')
resized2= cv2.resize(img_girl, (500, 500))
plt.imshow(resized2)
plt.title('Image')
plt.show()
```

```python
img_human = mpimg.imread('humsn.jpg')
resized3= cv2.resize(img_human, (500, 500))
plt.imshow(resized3)
plt.title('Image')
plt.show()
```

```python
blend = cv2.addWeighted(resized3,0.4,resized,0.7,0)
plt.imshow(blend)
plt.title('Blended Image')
plt.show()
```
